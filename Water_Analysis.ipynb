<font color="brown"><b><u><h2>IMPORTING LIBRARIES</h2></u></b></font>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<font color="brown"><b><u><h2>READING THE DATA FILE</h2></u></b></font>



```python
d=pd.read_csv("Final_Data.csv")
d
```

```python
#creating a list with all the states
states=["KARNATAKA","TAMILNADU","KERALA","PONDICHERRY","ANDHRA PRADESH"]
```

<font color="brown"><b><u><h2>ANALYSIS</h2></u></b></font>

<font color="red"><b>Q1.Which State Has the Highest number of water Bodies(rivers and lakes)?</b></font>

```python
state_lake_counts = d['STATE'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(10, 6))
state_lake_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Water Bodies(Rivers and Lakes) by Location:')
plt.ylabel('')
plt.show()
```

<strong><u>Conclusion:</u></strong><br>
<font color="blue">
Among AP, KA, KL, PD and TN<br> <strong><u>Andhra Pradesh</u></strong> has the highest number of water bodies(rivers and lakes).
</u></font>

<font color="brown"><u><h2>**TEMPERATURE**</h2></u></font>

<font color="red"><b>Q2.Average temperature in each state?</b></font>


```python
StateAvgtemp = {}
for index, row in d.iterrows():
    if row['STATE'] not in StateAvgtemp:
        StateAvgtemp[row['STATE']] = {'count': 0, 'sum': 0}
    if not pd.isna(row['TEMPERATURE °C : Mean']):
        StateAvgtemp[row['STATE']]['count'] += 1
        StateAvgtemp[row['STATE']]['sum'] += row['TEMPERATURE °C : Mean']

# Calculate the average temperature by dividing the sum by the count
for state, value in StateAvgtemp.items():
    StateAvgtemp[state] = value['sum'] / value['count']

# Plot the average temperature for each state
plt.figure(figsize=(10, 6))
plt.bar(list(StateAvgtemp.keys()), list(StateAvgtemp.values()))
plt.xticks(rotation=20)
plt.xlabel("Locaiton")
plt.ylabel("Avg. Temp")
plt.title("Average Temperature in Various Locations:")
plt.show()
```

<u>**Conclusion:**</u><br>
<font color="blue">The graph represents the average temperature of any given water body(rivers and lakes) in that particular state.<br>
Hence, from the graph we can conclude that <u><strong>Pondicherry</strong></u> has the highest average temperature among AP, KA, KL, PD, TN.
</u></b></font>


<font color="brown"><b><u><h2>pH Levels</h2></u></b></font>

<font color="red"><b>Q3.Are rivers Acidic,Basic or Nuetral?</b></font>

```python
StateAvgtemp = {}
for index, row in d.iterrows():
    if row['STATE'] not in StateAvgtemp:
        StateAvgtemp[row['STATE']] = {'count': 0, 'sum': 0}
    if not pd.isna(row['pH : Mean : 6.5-8.5']):
        StateAvgtemp[row['STATE']]['count'] += 1
        StateAvgtemp[row['STATE']]['sum'] += row['pH : Mean : 6.5-8.5']

# Calculate the average temperature by dividing the sum by the count
for state, value in StateAvgtemp.items():
    StateAvgtemp[state] = value['sum'] / value['count']

# Plot the average temperature for each state
plt.figure(figsize=(10, 6))
plt.bar(list(StateAvgtemp.keys()), list(StateAvgtemp.values()),color="green")
plt.xticks(rotation=35)
plt.xlabel("Locations")
plt.ylabel("Avg. pH")
plt.title("Average pH in Various Locations:")
plt.show()
```

<u>**Conclusion:**</u><br>
<font color="blue">
All states have average **pH=7.0**, from which we can conclude that all the water bodies(rivers and lakes) are neither acidic or basic.<br> <strong><u>Karnataka</u></strong> water bodies have slightly higher pH than other states.</font>

<font color="brown"><b><u><h2>BOD(Biological Oxygen Demand) and DO(Dissolved Oxygen) Levels</h2></u></b></font>

<font color="red"><b>Q4.Is the available Oxygen in water sufficient to meet the demands of aquatic life?</b></font>

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(d['STATE'], d['B.O.D. (mg/l) : Mean : < 3 mg/l'], color='blue')
plt.title('Average BOD Levels among various Locations:')
plt.xlabel('Locations')
plt.ylabel('Average BOD')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.bar(d['STATE'], d['D.O. (mg/l) : Mean : > 4 mg/l'], color='green')
plt.title('Average DO Levels among various Locations:')
plt.xlabel('Locations')
plt.ylabel('Average DO')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

<font color="red"><b>Q5.Is the available Oxygen in water sufficient to meet the demands of aquatic life in each state?</b></font>

```python
#Function to create lineplot for BOD and DO
def linplot(d,state,ticks=0):
  num_rows = len(d)
  rname_list = []
  for i in range(1, num_rows + 1):
    rname = 'R' + str(i)
    rname_list.append(rname)
  d['Rname'] = rname_list
  fig, ax = plt.subplots(figsize=(15, 8))
  x = d["Rname"]

  # Plot the first graph (lineplot)
  sns.lineplot(x=x, y=d["B.O.D. (mg/l) : Mean : < 3 mg/l"], ax=ax, label='Demand for Oxygen (BOD)',marker="o")

  # Plot the second graph (line plot)
  sns.lineplot(x=x, y=d["D.O. (mg/l) : Mean : > 4 mg/l"],ax=ax, label='Available Oxygen (DO)',marker="o")

  # Customize the plot
  ax.set_xlabel('Rivers')
  ax.set_ylabel('Oxygen')
  title="BOD V/S DO for "+state+":"
  ax.set_title(title)
  ax.grid(True)
  plt.legend(bbox_to_anchor=(1.05, 1), loc=1)
  plt.xticks(rotation=ticks)
  plt.show()
```

```python
linplot(d[d['STATE'] == states[0]],states[0])
```

```python
linplot(d[d['STATE'] == states[1]],states[1])
```

```python
linplot(d[d['STATE'] == states[2]],states[2])
```

```python
linplot(d[d['STATE'] == states[3]],states[3])
```

```python
linplot(d[d['STATE'] == states[4]],states[4],90)
```

<u>**Conclusion**</u>

<ol>
<li>Andra Pradesh has High demand for Oxygen while it has less dissolved Oxygen so the BOD has to be icreased in the water bodies by removing unwanted plants.</li>
<li>On the other hand, a balance is maintained between BOD and DO in other states.</li>
</ol>

<font color="brown"><h2><u>**Nitrate Levels**</u></h2></font>

<font color="red"><b>Q6.How does Nitrate levels vary from state to state?</b></font>

```python
StateAvgtemp = {}
for index, row in d.iterrows():
    if row['STATE'] not in StateAvgtemp:
        StateAvgtemp[row['STATE']] = {'count': 0, 'sum': 0}
    if not pd.isna(row['NITRATE- N+ NITRITE-N (mg/l) : Mean']):
        StateAvgtemp[row['STATE']]['count'] += 1
        StateAvgtemp[row['STATE']]['sum'] += row['NITRATE- N+ NITRITE-N (mg/l) : Mean']

# Calculate the average temperature by dividing the sum by the count
for state, value in StateAvgtemp.items():
    StateAvgtemp[state] = value['sum'] / value['count']

# Plot the average temperature for each state
plt.figure(figsize=(10, 6))
plt.bar(list(StateAvgtemp.keys()), list(StateAvgtemp.values()),color="green")
plt.xticks(rotation=35)
plt.xlabel("Locations")
plt.ylabel("Avg. nitrate")
plt.title("Average nitrate value in Various Locations:")
plt.show()
```

<u>**Conclusion**</u><br>
<font color="blue">
The average nitrate value is over 2.5 mg/L in <u>**Andra Pradesh**</u>, and is the highest Nitrate levels among the following.
<br>
This followed by Karnataka, Kerela, Tamil Nadu, Pondicherry.

<font color="brown"><b><u><h2>CONDUCTIVITY</h2></u></b></font>

<font color="red"><b>Q7. How does conductivity effect water bodies?</b></font>

```python
StateAvgtemp = {}
for index, row in d.iterrows():
    if row['STATE'] not in StateAvgtemp:
        StateAvgtemp[row['STATE']] = {'count': 0, 'sum': 0}
    if not pd.isna(row['CONDUCTIVITY: Mean']):
        StateAvgtemp[row['STATE']]['count'] += 1
        StateAvgtemp[row['STATE']]['sum'] += row['CONDUCTIVITY: Mean']

# Calculate the average conductivity value by dividing the sum by the count
for state, value in StateAvgtemp.items():
    StateAvgtemp[state] = value['sum'] / value['count']

# Plot the average conductivity for each state
plt.figure(figsize=(10, 6))
plt.bar(list(StateAvgtemp.keys()), list(StateAvgtemp.values()),color="green")
plt.xticks(rotation=35)
plt.xlabel("Locations")
plt.ylabel("Average Conductivity")
plt.title("Average conductivity value in Various Locations:")
plt.show()
```

<u>**Conclusuion**</u><br>
<font color="blue"><u>**Pondicherry**</u> has the highest Conductivity
According to the WHO reports higher the Conductivity more the chances of presence of pollutants. It does not specify the type of Pollutant but helps to conclude the presence of pollutants.
